# Simple Boolean masking

    1. We can filter a dataframe by using a boolean series called as mask.
    
    2. To create a mask we can use inequality operators.
    
    3. We can then pass this boolean series(mask) into the dataframe and get the filtered rows.

```python
import numpy as np
import pandas as pd
```

```python
pokemon = pd.read_csv('datasets/pokemon.csv')
pokemon['Length'] = pokemon['Name'].apply(len)
pokemon
```

    Q. Filter out all the pokemons which have the type as 'Grass, Poison'.

```python
mask = pokemon['Type'] == 'Grass, Poison'
mask # This returns us a series
```

    We can now pass this series into the dataframe to get a new dataframe with the filtered rows.

```python
grass_poison = pokemon[mask]
grass_poison
```

# & Operator

    Q. Find all the pokemons whose name is atleast 9 characters long and which are of the type 'Grass, Poison'.

```python
import numpy as np
import pandas as pd
```

```python
pokemon = pd.read_csv('datasets/pokemon.csv')
pokemon['Length'] = pokemon['Name'].apply(len)
pokemon
```

```python
mask1 = pokemon['Length'] >= 9
mask2 = pokemon['Type'] == 'Grass, Poison'
```

```python
mask1
```

```python
mask2
```

```python
mask = mask1 & mask2
mask
```

```python
ans = pokemon[mask]
ans
```

# The isin() method

    The isin() method is used to return a boolean series(mask) which then can be used to pass it into the 
    dataframe to filter out the rows.

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.read_csv('datasets/employees.csv')
employees
```

    Q. We want to find all the men who are in Finance or Product or Distribution 

```python
mask1 = employees['Gender'] == 'Male'
```

```python
mask2 = employees['Team'].isin(['Finance', 'Product', 'Distribution'])
mask2
```

```python
mask = mask1 & mask2 # as we want to find all the men in finance, product or distribution
mask
```

```python
ans = employees[mask]
ans # This is the list of all the men who are in Finance, Product or Distribution.
```

# isnull() and notnull() methods

    1. The isnull() and notnull() methods are used to return a boolean series(mask) which then can be passed to a 
       dataframe to filter out the rows of our concern.
    
    2. The isnull() method gives a True value wherever it finds a null value.
    
    3. The notnull() method gives a False value wherever it finds a null value. It is the exact opposite of isnull() method.

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.read_csv('datasets/employees.csv')
employees
```

    Q. Find all people who do not have first name but are working in a department 🙄🙄🙄.

```python
mask1 = employees['First Name'].isnull() # rows where First Name is null
mask2 = employees['Team'].notnull() # rows where Team is not null.

mask = mask1 & mask2
employees[mask]
```
