### title, xlabel, ylabel

```python
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.show()
```

### label and legend
    The possible values for legend loc are :
    'best', 
    'upper right', 
    'upper left', 
    'lower left', 
    'lower right', 
    'right', 
    'center left', 
    'center right', 
    'lower center', 
    'upper center', 
    'center'
    
    The default value is 'best' which means that matplotlib engine will automatically determine the best position according
    to the available space in your plot.

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend(loc = 'best')
plt.show()
```

### color

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### marker

    , 	pixel
    o	circle
    v	triangle_down
    ^	triangle_up
    <	triangle_left
    >	triangle_right
    1	tri_down
    2	tri_up
    3	tri_left
    4	tri_right
    8	octagon
    s	square
    p	pentagon
    P	plus (filled)
    *	star
    h	hexagon1
    H	hexagon2
    +	plus
    x	x
    X	x (filled)
    D	diamond
    d	thin_diamond
    |	vline
    _	hline
    0 (TICKLEFT)	tickleft
    1 (TICKRIGHT)	tickright
    2 (TICKUP)	tickup
    3 (TICKDOWN)	tickdown
    4 (CARETLEFT)	caretleft
    5 (CARETRIGHT)	caretright
    6 (CARETUP)	caretup
    7 (CARETDOWN)	caretdown
    8 (CARETLEFTBASE)	caretleft (centered at base)
    9 (CARETRIGHTBASE)	caretright (centered at base)
    10 (CARETUPBASE)	caretup (centered at base)
    11 (CARETDOWNBASE)	caretdown (centered at base)


```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = '$s$')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue', marker = 9)
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### linestyle
      Solid line

    -- Dashed line

    -. dash-dot line

    :  Dotted line
    
    The supported values are :
    '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's', linestyle = 'dashed')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue', marker = 'o', linestyle = '-.')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### linewidth

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', \
         color = 'blue', marker = 'o', linestyle = '-.', linewidth = 2) # lw
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### grid

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', \
         color = 'blue', marker = 'o', linestyle = '-.', linewidth = 2) # lw
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.grid()
plt.show()
```

### Limiting the axes

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', \
         color = 'blue', marker = 'o', linestyle = '-.', linewidth = 2) # lw
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.grid()
plt.ylim(50, 700) # limiting the axes
plt.show()
```

# scatter plots

    1. Used for bivariate analysis
    2. Numerical vs Numerical analysis only
    3. Used to study correlation between data

```python
x = np.linspace(-10, 10, 50)
y = 10 * x + 3 * np.random.randint(0, 301, 50)
```

```python
plt.scatter(x, y)
plt.title('x vs y')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
df = pd.read_csv('datasets/batter.csv')
```

```python
df = df[:50]
```

```python
df
```

```python
plt.scatter(df['avg'], df['strike_rate'])
plt.title('Avg Vs Strike rate of top 50 batsman')
plt.xlabel('Avg')
plt.ylabel('Strike rate')
plt.tight_layout()
plt.savefig(fname = 'Img.png')
plt.show()
```

```python
plt.scatter(df['avg'], df['strike_rate'], color = 'red', marker = '+')
plt.title('Avg Vs Strike rate of top 50 batsman')
plt.xlabel('Avg')
plt.ylabel('Strike rate')
plt.tight_layout()
plt.savefig(fname = 'Img.png')
plt.show()
```

### size

```python
plt.scatter(df['avg'], df['strike_rate'], color ='salmon', marker = '+', s = df['runs']/50)
plt.title('Avg Vs Strike rate of top 50 batsman')
plt.xlabel('Avg')
plt.ylabel('Strike rate')
plt.tight_layout()
plt.savefig(fname = 'Img.png')
plt.show()
```

# plt.plot vs plt.scatter

```python
# We can draw scatter plots using the plt.plot()
# This is faster than plt.scatter() 
plt.plot(df['avg'], df['strike_rate'], '+')
plt.title('Avg Vs Strike rate of top 50 batsman')
plt.xlabel('Avg')
plt.ylabel('Strike rate')
plt.tight_layout()
plt.savefig(fname = 'Img.png')
plt.show()
```

# bar chart

    Used for bivariate analysis
    used for categorical data

```python
colors = ['red', 'blue', 'green', 'yellow', 'pink']
children = [10, 20, 30, 40, 50]

plt.bar(colors, children, color = 'salmon', width = 0.3)
plt.xlabel('colors')
plt.ylabel('# of children')
plt.title('# of children vs colors')
plt.xticks(rotation = 45)
plt.show()
```

# pie chart

```python
names = ['Abhishek', 'Sameer', 'Siddharth', 'Ibrahim', 'Viraj', 'karan']
marks = [10,24,45,67,89,90]
```

```python
plt.pie(marks)
plt.show()
```

```python
plt.pie(marks, labels = names)
plt.show()
```

```python
plt.pie(marks, labels = names, colors = ['red', 'blue', 'green', 'yellow', 'pink', 'salmon'])
plt.show()
```

```python
plt.pie(marks, labels = names, \
        colors = ['red', 'blue', 'green', 'yellow', 'pink', 'salmon'], \
        autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(marks, labels = names, \
        colors = ['red', 'blue', 'green', 'yellow', 'pink', 'salmon'], \
        autopct = '%0.1f%%', \
        explode = [0, 0, 0, 0, 0.1, 0.4]
       )
plt.show()
```
