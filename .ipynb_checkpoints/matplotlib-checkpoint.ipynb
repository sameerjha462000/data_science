```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

    The line plot is used for bivariate data.
    

```python
prices = [45000, 55000, 58000, 47000, 43000]
years = [2015, 2016, 2017, 2018, 2019]

plt.plot(years, prices)
plt.show()
```

```python
batsman = pd.read_csv('datasets/sharma-kohli.csv')
```

```python
batsman
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.show()
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
plt.show()
```

### title, xlabel, ylabel

```python
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.show()
```

### label and legend
    The possible values for legend loc are :
    'best', 
    'upper right', 
    'upper left', 
    'lower left', 
    'lower right', 
    'right', 
    'center left', 
    'center right', 
    'lower center', 
    'upper center', 
    'center'
    
    The default value is 'best' which means that matplotlib engine will automatically determine the best position according
    to the available space in your plot.

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend(loc = 'best')
plt.show()
```

### color

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### marker

    v	triangle_down

    >	triangle_right
    1	tri_down
    2	tri_up

    4	tri_right
    8	octagon
    s	square
    p	pentagon

    h	hexagon1
    H	hexagon2

    X	x (filled)

    d	thin_diamond
    |	vline
    _	hline
    0 (TICKLEFT)	tickleft
    1 (TICKRIGHT)	tickright
    2 (TICKUP)	tickup
    3 (TICKDOWN)	tickdown
    4 (CARETLEFT)	caretleft
    5 (CARETRIGHT)	caretright
    6 (CARETUP)	caretup
    7 (CARETDOWN)	caretdown

    9 (CARETRIGHTBASE)	caretright (centered at base)

    11 (CARETDOWNBASE)	caretdown (centered at base)


```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = '$s$')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue', marker = 9)
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### linestyle
      Solid line

    -- Dashed line

    -. dash-dot line

    :  Dotted line
    
    The supported values are :
    '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's', linestyle = 'dashed')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', color = 'blue', marker = 'o', linestyle = '-.')
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```

### linewidth

```python
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Rohit Sharma', color = 'red', marker = 's')
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat Kohli', \
         color = 'blue', marker = 'o', linestyle = '-.', linewidth = 2) # lw
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('years')
plt.ylabel('prices')
plt.legend()
plt.show()
```
